```python
import matplotlib.pyplot as plt
df = sns.load_dataset("tips")
df
```

```python
sns.scatterplot(data = df,x = 'total_bill',y = 'tip',hue = 'day')


plt.title("Scatter plot : Total Bill vs Tip")
plt.show()
```

```python
sns.lineplot(data=df, x='total_bill', y='tip')

plt.title("Line Plot: Total Bill vs Tip")
plt.show()
```

```python
sns.barplot(x = 'day', y = 'total_bill', data = df)
plt.title("Average Total bill by day")
plt.show()
```

```python
sns.boxplot(data = df , x = 'day', y = 'total_bill',hue = 'day')
plt.title("Box plot : total bill  distribution by day")
plt.show()
```

```python
df = sns.load_dataset("tips")

sns.kdeplot(df["total_bill"],shade = True,color = 'blue')
plt.title("kernal density estimate (KDE) of total bill")
plt.xlabel("Total bill")
plt.ylabel("Density")
plt.show()
```

```python
sns.violinplot(x = 'day',y = 'total_bill',data = df)
plt.title("violin plot : total bill distribution by dAY")
plt.show()
```

```python
sns.histplot(df['total_bill'],kde =True)
plt.title("Distribution plot:total Bill distribution by day")
plt.show()
```

```python
sns.pairplot(df,hue='sex')
plt.title("pairwise plot : tips dataset")
plt.show()
```

```python
df = sns.load_dataset("tips")
corr = df.corr(numeric_only=True)
sns.heatmap(corr,annot = True,cmap='coolwarm')
plt.show()
```
