```python
import numpy as np
x = np.array([1,2,3,4,5]) #1D ARRAY
print(x)
b = x[1:2] #Acess 2
print(b)
print(x[-3:-1])
print(x[1:5:2])
```

```python
y = np.array([[1,2,3],
              [4,5,6]])#2D ARRAY
print(y)
a = y[0,1:2] # Access 2
print(a)
```

```python
z = np.array([[[1,2,3],
             [4,5,6]],
            [[5,6,7],
             [4,5,6]]])#3D ARRAY
print(z)
c = z[1,1,1] #Access 5
print(c)
print(z.ndim) #To know the dimensions 
```

```python
a = np.array([1,2,3,4])
print(a.dtype)
```

```python
a = np.array([1,2,3,4],dtype='S')
print(a)
print(a.dtype)
```
