```python
import numpy as np
import pandas as pd
```

```python
data = [
[2001,"C101","2023-08-01 18:00","2023-08-01 18:45",500,"upi","Dominos",4],
[2002,"C102","2023-08-01 19:00","2023-08-01 18:50",700,"Credit card","KFC",5],
[2003,"C103","2023/08/02 20:00",None,None,"Debit Card","McDonalds",3],
[2004,"C104","invalid","2023-08-03 21:00",800,"UPI","Dominos",6],
[2005,"C105","2023-08-04 18:30","2023-08-04 19:00",-200,"Cash","KFC",2],
[2005,"C105","2023-08-04 18:30","2023-08-04 19:00",-200,"Cash","KFC",2],
[2006,"C106","2023-08-05 17:00","2023-08-05 17:40",1200,None,"Pizza Hut",4]
]

columns = ["order_id","customer_id","order_time","delivery_time","order_value","payment_method","restaurant","rating"]

df = pd.DataFrame(data, columns=columns)
df
```

```python
#Cleaning
print(df.info())
```

```python
print(df.isnull().sum)
```

```python
print(df.duplicated().sum())
```

```python
#1Fix date and time coluumns
df['order_time'] = pd.to_datetime(df['order_time'], errors='coerce')
df['delivery_time'] = pd.to_datetime(df['delivery_time'], errors='coerce')
```

```python
# Remove inalid rec
#drop - only few null or fill - mostly null

#Drop rows where order_time and del time is missing
#df = df[df['order_time'].dropna()]



#df = df[df['delivery_time'].dropna()]

#OR

df = df.dropna(subset=['order_time','delivery_time'])

df
```

```python
#2remove rows where deltime is before order
df = df[df['delivery_time'] >= df['order_time']]
df

```

```python
#4Handling incorrect val
df = df[df['order_value'] > 0]


#fill missing val
df['payment_method'].fillna('unknown',inplace = True)
```

```python
df
```

```python
#5
df = df[(df['rating'] >= 1) & (df['rating'] <= 5)]
df
```

```python
df.drop_duplicates(inplace=True)
```

```python
#6
df['payment_method'] = df['payment_method'].str.upper()

df['restaurant']=df['restaurant'].str.title()
df
```

```python
#validation checks
assert df.duplicated().sum()==0
assert (df['order_value']>0).all()
```

```python
df.to_csv("orders_food_cleaned.csv",index=False)
```
